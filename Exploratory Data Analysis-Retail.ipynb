#### 5.Exploratory Data Analysis

#### Statewise analysis of Profit Discount and Sales

```python
df_state= df.groupby(['State'])[['Sales', 'Discount', 'Profit']].mean()# grouping state w.r to mean of each column
df_state.head(10)
```

#### Statewise Profit Analysis

```python
df_state1=df_state.sort_values('Profit')
print(df_state1)
```

```python
df_state1[['Profit']].plot(kind = 'bar', figsize = (15,4))
plt.title('State wise Profit Analysis', fontsize = 20)
plt.ylabel('Profit')
plt.xlabel('States')
plt.show()
```

#### Inference:
    Vermont: Highest Profit
    Ohio: Lowest Profit

```python
df_state2=df_state.sort_values('Sales')
df_state2['Sales'].plot(kind='bar',figsize = (15,4))
plt.title('State wise analysis of Sale',fontsize=20)
plt.xlabel('States')
plt.ylabel('Sales')
```

#### Inference:
         Wyoming : Highest Sales
    South Dakota : Minimum Sales

```python
df_state3=df_state.sort_values('Discount')
df_state3['Discount'].plot(kind='bar',figsize=(15,4))
plt.title('State wise analysis of Discount', fontsize=20)
```

#### Inference
     No Discount was given to 21 States like Alabama,Nebraska,Georgia,etc
     Highest dicount is given to Illinois

#### Category wise Sales,Profit and Discount

```python
df_category = df.groupby(['Category'])[['Sales', 'Quantity', 'Discount', 'Profit']].mean()
df_category.head()
```

#### Sub-Category wise Sales, Profit and Discount

```python
df_sub_category = df.groupby(['Sub-Category'])[['Sales', 'Quantity', 'Discount', 'Profit']].mean()
df_sub_category.head(10)
```

```python
df_sub_category1=df_sub_category.sort_values('Profit')
df_sub_category1['Profit'].plot(kind='bar',figsize=(15,6),color="Purple")
plt.title('Sub-Category wise analysis of Profit', fontsize=20)
plt.ylabel('Profit')
```

#### Inference:
    Copiers: High Profit
    Tables : Negative/No Profit

```python
plt.figure(figsize = (12,12))
plt.pie(df_sub_category['Sales'], labels = df_sub_category.index, autopct = '%1.1f%%')
plt.title('Sub-Category Wise Sales Analysis', fontsize = 20)
plt.legend()
plt.xticks(rotation = 90)
plt.show()
```

#### Inference:
    Copiers   : Highest Sales
    Fasteners : Minimum Sales

```python
plt.figure(figsize = (12,12))
plt.pie(df_sub_category['Discount'], labels = df_sub_category.index, autopct = '%1.1f%%')
plt.title('Sub-Category Wise Discount Analysis', fontsize = 20)
plt.legend()
plt.xticks(rotation = 90)
plt.show()
```

#### Inference:
    Binders : High Discount
    Labels  : Less Dicount
